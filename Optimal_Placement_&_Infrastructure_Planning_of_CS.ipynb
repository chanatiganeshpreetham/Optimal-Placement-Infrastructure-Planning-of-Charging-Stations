```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data_set = pd.read_csv('data36.csv')
```

```python
df = pd.DataFrame(data_set)
r1=6.5 # weightage of type 1 POI
r2=4
r3=2
r4=1
pp1=0.5
pp2=0.3
pp3=0.2
pp4=0.1
pr1=0.03 #3%(probability of random vehicle taking chrage on type 1 road)3% VR_typ1 means vehicles roaming on type1 road in each 1hr
pr2=0.02 #2%
pr3=0.01 #1%

# Define the equation
def rankpoiequation(row):
    return r1 *pp1* row['POI_1'] + r2 *pp2* row['POI_2'] + r3*pp3 * row['POI_3'] +r4 *pp4* row['POI_4'] 

# Add a new column to the data
df['RankPOI'] = df.apply(rankpoiequation, axis=1)

#print(df)

# Define the equation
def rankroamingequation(row):
    return pr1 * row['Roaming_vehicles_in_type1'] + pr2 * row['Roaming_vehicles_in_type2'] + pr3 * row['Roaming_vehicles_in_type3'] 

# Add a new column to the data
df['Rankroaming'] = df.apply(rankroamingequation, axis=1)

# Define the equation
def rank(row):
    return row['RankPOI'] +row['Rankroaming']
df['Rank'] = df.apply(rank, axis=1)
print(df)
```
